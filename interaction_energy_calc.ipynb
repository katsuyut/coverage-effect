```python
D = 0.122 # D
uconv1 = 3.33564e-30 # C·m/D
D = D*uconv1 # C·m
D
```

```python
h = 1.13E-10 # m
q = D/h # C
q
```

```python
k = 8.99E+09 # N·m2·C−2
uconv2 = 6.24E+18 # eV/J
```

```python
xrange = np.linspace(1.5,5)*1e-10
```

```python
V = 2 * k * q**2 * ((1/xrange - 1/(xrange**2 + h**2)**0.5)) * uconv2
```

```python
plt.plot(xrange, V)
plt.xlabel('distance [m]')
plt.ylabel('Energy [eV]')
plt.show()
```
