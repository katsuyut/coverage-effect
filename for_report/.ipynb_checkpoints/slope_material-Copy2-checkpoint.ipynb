```python
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from sklearn.metrics import r2_score
```

```python
def MSE(a, b):
    mse = 0
    mse += sum((a - b)**2)/len(a)
    return mse
```

```python
alpha = [0.06, 0.05, 0.04, 0.03, 0.02, 0.01, 0.005, 0.002, 0.001]
blis = []
mselis = []
r2lis = []
for al in alpha:
    print('For {}'.format(al))
    lasso = linear_model.Lasso(alpha=al)
    sscaler = StandardScaler()

    sscaler.fit(mergedadd.T[1:].T)
    X = sscaler.transform(mergedadd.T[1:].T)
    lasso.fit(X, mergedadd['averaged'])
    print(lasso.intercept_)
    prediction = lasso.predict(X)
    mse = MSE(prediction, dfadd['averaged'])
    r2 = r2_score(prediction, dfadd['averaged'])
    b = 0
    for i in range(len(lasso.coef_)):
        if lasso.coef_[i] != 0:
            print(mergedadd.columns[i+1])
            b += 1
    print('b', b)
    print('MSE', mse)
    print('R2', r2)
    blis.append(b)
    mselis.append(mse)
    r2lis.append(r2)
    print( )
```

```python
plt.rcParams["font.size"] = 20
# fig = plt.figure(figsize=(5,4))

plt.plot(alpha, blis)
plt.xlabel(r'$\alpha$')
# plt.xlabel('alpha')
plt.ylabel('Number of properties')
plt.savefig('lassonumofprop.eps', format='eps', bbox_inches="tight")
plt.show()
```

```python
# fig = plt.figure(figsize=(5,4))
plt.plot(alpha, mselis)
plt.xlabel(r'$\alpha$')
# plt.xlabel('alpha')
plt.ylabel('MSE')
plt.savefig('lassomse.eps', format='eps', bbox_inches="tight")
plt.show()
```

```python
# fig = plt.figure(figsize=(5,4))
plt.plot(alpha, r2lis)
plt.xlabel(r'$\alpha$')
# plt.xlabel('alpha')
plt.ylabel('$R^2$ score')
plt.ylim(0,1)
plt.savefig('lassor2.eps', format='eps', bbox_inches="tight")
plt.show()
```
