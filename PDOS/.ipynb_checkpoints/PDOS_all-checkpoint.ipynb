```python
%matplotlib inline
import numpy as np
from pymatgen.io.vasp.outputs import Vasprun
from pymatgen.electronic_structure.plotter import DosPlotter
from pymatgen.electronic_structure.core import Spin
# import matplotlib.pyplot as plt
# import pymatgen.electronic_structure.dos
```

```python
# v = Vasprun("./Cu_111_u2/vasprun.xml")
# v = Vasprun("./Cu_111_u2_no01_CO_n1_d8/vasprun.xml")
v = Vasprun("./Cu_111_u2_no08_CO_n1_d8/vasprun.xml")
tdos = v.tdos
cdos = v.complete_dos
structure = cdos.structure
```

TDOS

```python
plotter = DosPlotter()
plotter.add_dos("Total DOS", tdos)
plotter.show(xlim=[-15, 10], ylim=[0, 100])
```

PDOS

spd_dos

element_dos

element_spd_dos

```python
spd_dos = cdos.get_spd_dos()
plotter = DosPlotter()
plotter.add_dos_dict(spd_dos)
plotter.show(xlim=[-15, 10], ylim=[0, 100])

element_dos = cdos.get_element_dos()
plotter = DosPlotter()
plotter.add_dos_dict(element_dos)
plotter.show(xlim=[-15, 10], ylim=[0, 100])
```

```python
for ele in ['Cu', 'C', 'O']:
    element_spd_dos = cdos.get_element_spd_dos(ele)
    plotter = DosPlotter()
    plotter.add_dos_dict(element_spd_dos)
    plotter.show(xlim=[-15, 10], ylim=[0, 100])
```

```python
plotter = DosPlotter()
for isite,site in enumerate(structure[12:16]):
    site_dos = cdos.get_site_dos(site)
    plotter.add_dos('Cu'+str(isite+1), site_dos)
plotter.show(xlim=[-15, 10], ylim=[0, 10])
```

```python
plotter = DosPlotter()
for isite,site in enumerate(structure[12:14]):
    site_spd_dos = cdos.get_site_spd_dos(site)
    plotter.add_dos_dict(site_spd_dos)
    plotter.show(xlim=[-15, 10], ylim=[0, 10])
```
